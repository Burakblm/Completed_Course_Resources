## MNİST VERİ KÜMESİNİ YÜKLEMEK

```python
from keras.datasets import mnist
(train_images,train_labels),(test_images,test_labels) = mnist.load_data()
```

# Yapay Sinir Ağı Mimarisi

```python
from keras import models
from keras import layers
```

```python
networks = models.Sequential()
networks.add(layers.Dense(512,activation="relu",input_shape=(28*28,)))
networks.add(layers.Dense(10,activation="softmax"))
```

### YSA Modelinin derlenmesi

```python
networks.compile(optimizer="rmsprop",
                 loss="categorical_crossentropy",
                 metrics = ["accuracy"])
```

### Girişlerin Ve Etiketlerin Hazırlanması

```python
train_images = train_images.reshape((60000,28*28))
train_images = train_images.astype('float32') / 255

test_images = test_images.reshape((10000,28*28))
test_images = test_images.astype('float32') / 255
```

```python
from keras.utils.np_utils import to_categorical
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
print(train_images.shape)
print(train_labels.shape)
```

## YSA Modelinin Eğitilmesi

```python
networks.fit(train_images,train_labels,epochs=5,batch_size=128)
```

```python
test_loss , test_accuracy = networks.evaluate(test_images,test_labels)
```

```python
print("test loss: ",test_loss," test accuracy: ",test_accuracy)
```
