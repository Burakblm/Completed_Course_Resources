```python
# gerekli kütüphanelerin import edilmesi
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
#sigmoid fonksiyonu
def sigmoid(x):
    a = []
    for i in x:
        a.append(1/(1+math.exp(-i)))
    return a
```

```python
# ReLu fonksiyonu
def re(x):
    b = []
    for i in x:
        if i<0:
            b.append(0)
        else:
            b.append(i)
    return b
```

```python
# leaky ReLu fonksiyonu
def lr(x):
    b = []
    for i in x:
        if i<0:
            b.append(i/10)
        else:
            b.append(i)
    return b
```
